# 🔹 What is Machine Learning?

**Machine Learning (ML)** is a subset of **Artificial Intelligence (AI)** that enables systems to learn from data without explicit programming.

It improves over time as it gains more data.

## Types of Machine Learning

### 1. Supervised Learning
- Learns from labeled data
- Used for prediction and classification

### 2. Unsupervised Learning
- Learns from unlabeled data
- Used to find hidden patterns and relationships

### 3. Reinforcement Learning
- Learns through rewards and penalties
- Commonly used in robotics and gaming

---

# 🔹 Applications of Machine Learning

Machine Learning is used in many real-world applications:

- 🤖 Make robots recognize faces or learn to walk
- 📊 Discover patterns in user logs and financial data
- 👥 Segment customers for personalized marketing
- 🛒 Recommend products based on user behavior
- 🔒 Detect fraudulent transactions
- 📈 Forecast sales and demand
- ⚙️ Automate and optimize real-world decisions

# 🔹 Common Machine Learning Libraries

## Scikit-Learn
- Ideal for:
  - Regression
  - Classification
  - Clustering
- Beginner-friendly machine learning library

## TensorFlow
- Developed by Google
- Used for large-scale machine learning and deep learning

## Keras
- High-level deep learning API
- Comes integrated with TensorFlow
- Easy for beginners

> 💡 For deep learning beginners, start with **Keras**.

## PyTorch
- Developed by Meta (Facebook)
- Popular for deep learning research
- Easy to use and debug

## Pandas & NumPy

### Pandas
- Data cleaning
- Data manipulation
- Data analysis

### NumPy
- Numerical computing
- Array and matrix operations

---

# 🔹 PyTorch vs TensorFlow

| Feature | PyTorch | TensorFlow |
|----------|----------|------------|
| Developer | Meta (Facebook) | Google |
| Learning Curve | Easier | Moderate |
| Debugging | Easier | More Complex |
| Research Usage | Very Popular | Popular |
| Production Deployment | Good | Excellent |
| Community Support | Large | Large |

---

## ✅ When to Use PyTorch?

- Research projects
- Learning deep learning
- Rapid prototyping
- Easy debugging

---

## ✅ When to Use TensorFlow?

- Production systems
- Mobile deployment
- Enterprise applications
- Large-scale machine learning
