# Types of Experiments in Probability

---

## Types of Experiments

### Deterministic Experiment
- Always produces the same outcome when repeated under identical conditions  

### Probabilistic (Random) Experiment
- Outcome is uncertain  
- May vary even if repeated under identical conditions  

---

## Sample Space

The **sample space** is the set of all possible outcomes of a probabilistic experiment.

### Examples:

- Rolling a Die → `{1, 2, 3, 4, 5, 6}`  
- Tossing 2 Coins → `{(HH), (HT), (TH), (TT)}`  

---

## Number of Outcomes

### For Coins

Number of possible outcomes:

$$
2^n
$$

Where:
- **n** = number of coins  

---

### For Dice

Number of possible outcomes:

$$
6^n
$$

Where:
- **n** = number of dice
