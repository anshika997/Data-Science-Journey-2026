# 🤖 Using Different ML Models

Machine Learning (ML) is a branch of **Artificial Intelligence (AI)** that enables computers to learn from data instead of being explicitly programmed with rules.

---

### Example: Deep Blue

Deep Blue was a chess-playing computer developed by IBM.

It was extremely good at chess, but it could not:

- Recognize images
- Drive cars
- Detect spam emails

- Spam Detection

# 🔹 What Is a Classifier?

A **Classifier** is a Machine Learning model used to predict categories or classes.

It:

- Takes input data
- Learns patterns from examples
- Predicts labels for new data

This process is called **Supervised Learning**.

---

## Example

Input:

```text
Weight = 150g
Texture = Smooth
```

Output:

```text
Apple
```

The classifier learns how features relate to labels.

---

# 🔹 Features and Labels

Machine Learning models learn using:

## Features

Features are the characteristics or attributes of the data.

### Examples

- Weight
- Color
- Height
- Age
- Texture

Example:

```python
[140, 1]
```

where:

- 140 = Weight
- 1 = Texture

---

## Labels

Labels are the answers we want the model to predict.

### Examples

```text
Apple
Orange
Spam
Not Spam
Cat
Dog
```

---

# 🧠 Learning Algorithm vs Classifier

Think of it this way:

### Learning Algorithm

The process that learns patterns from data.

### Classifier (Model)

The final trained system that makes predictions.

```text
Training Data
      ↓
Learning Algorithm
      ↓
Trained Classifier (Model)
      ↓
Predictions
```

---

# 🔹 How Does It Work?

Let's train a machine to identify apples and oranges.

---

## Step 1: Import Decision Tree

```python
from sklearn import tree
```

---

## Step 2: Create Training Data

```python
features = [
    [140, 1],
    [130, 1],
    [150, 0],
    [170, 0]
]
```

### Meaning

| Weight | Texture |
|----------|----------|
| 140 | 1 |
| 130 | 1 |
| 150 | 0 |
| 170 | 0 |

---

## Step 3: Create Labels

```python
labels = [0, 0, 1, 1]
```

Where:

| Label | Fruit |
|---------|---------|
| 0 | Apple |
| 1 | Orange |

---

## Step 4: Train the Model

```python
clf = tree.DecisionTreeClassifier()

clf = clf.fit(features, labels)
```

### What Does fit() Mean?

```python
fit()
```

means:

> Learn patterns from the training data.

The model studies the relationship between features and labels.

---

## Step 5: Make Predictions

```python
print(clf.predict([[160, 0]]))
```

Output:

```python
[1]
```

Prediction:

```text
Orange
```

The model uses learned patterns to classify new data.

---

# 🔍 Understanding the Workflow

```text
Training Data
      ↓
fit()
      ↓
Learn Patterns
      ↓
Create Model
      ↓
Predict New Data
```
