# 🤖 How Machines Learn

Not so long ago, if you had picked up your phone, opened ChatGPT, and asked it what to do in a particular situation, it would have ignored you—and people might have looked at you like you were losing your mind. Today, it's completely normal.

Technology is growing faster than ever, and **Machine Learning (ML)** is at the center of this transformation.

---

# 📧 Early Example: Spam Filters

One of the earliest real-world applications of Machine Learning was the **spam filter**.

Back in the 1990s, email inboxes were often flooded with unwanted messages about:

- Dubious offers
- Miracle diets
- Suspicious job proposals

Engineers began building systems that could learn to recognize spam patterns from examples.

Instead of writing rules such as:

> "If the email contains the word 'lottery', mark it as spam."

they allowed the system to learn from examples of:

- Spam emails
- Non-spam (ham) emails

As users continued marking emails as spam, the system improved and learned to detect new spam messages automatically.

This was an early example of Machine Learning in action.

---

# ❌ What Machine Learning Is Not

If you download hundreds of books onto your computer, it does **not** automatically become intelligent.

The computer:

- Does not start writing like Shakespeare
- Does not summarize books automatically
- Does not understand the content

It simply stores information.

This is called **data storage and retrieval**, not Machine Learning.

---

# ✅ What Machine Learning Is

Machine Learning is the science (and art) of programming computers so they can learn from data.

### Arthur Samuel's Definition (1959)

> "Machine learning is the field of study that gives computers the ability to learn without being explicitly programmed."

### Tom Mitchell's Definition (1997)

> "A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E."

---

# 🔍 Understanding the Definition

Let's use the spam filter example.

### Task (T)

Identify whether an email is:

- Spam
- Not Spam (Ham)

### Experience (E)

A collection of labeled emails:

- Spam emails
- Non-spam emails

### Performance Measure (P)

Accuracy of classification.

For example:

- Correctly identifying spam emails
- Correctly identifying legitimate emails

If the spam filter improves as it sees more emails, it is learning.

# 📚 Training Set and Training Instances

The examples used to teach a Machine Learning system are called the:

## Training Set

A collection of data used to train the model.

### Example

| Email | Label |
|---------|---------|
| Win a free iPhone | Spam |
| Meeting at 2 PM | Ham |
| Claim your prize now | Spam |

---

## Training Instance (Sample)

A single example from the training set.

Example:

| Email | Label |
|---------|---------|
| Win a free iPhone | Spam |

This single row is a training instance.

---

# 🧠 What Is a Model?

The part of a Machine Learning system that learns patterns and makes predictions is called the **Model**.

Think of the model as the brain of the ML system.

After training, the model can predict outcomes for new data.

---

# ⚙️ Traditional Programming vs Machine Learning

To understand ML better, compare it with traditional programming.

---

## Traditional Programming

In traditional programming:

A human writes explicit rules for the computer.

### Formula

```text
Input + Rules = Output
```

### Example

Rule:

```text
If email contains "free money"
→ Mark as Spam
```

Problem:

Spammers can change words:

```text
frëe m0ney
```

The rule may fail.

---

## Machine Learning Approach

Instead of writing rules manually:

The computer learns patterns from examples.

### Training Phase

```text
Input Data + Correct Outputs
            ↓
         Learning
            ↓
          Model
```

### Prediction Phase

```text
New Input + Model
         ↓
     Prediction
```

---

# 🎯 Why Machine Learning Is Powerful

Machine Learning can:

- Learn hidden patterns
- Adapt to changing data
- Improve with experience
- Handle situations not explicitly programmed

For example:

A spam filter can recognize spam messages it has never seen before by learning general spam characteristics.

---

# 📋 Summary

## Key Points

- Simply storing data is **not** Machine Learning.
- ML is about improving performance through experience.
- Systems learn patterns from data instead of following hardcoded rules.
- The trained component is called the **Model**.
- Spam detection was one of the earliest successful ML applications.
- ML systems adapt and improve as they receive more data.
- Traditional Programming uses rules written by humans.
- Machine Learning learns those rules automatically from data.
