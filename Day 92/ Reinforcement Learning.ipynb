# 🎮 Reinforcement Learning

Imagine a small robot standing in front of a maze.

Its goal is simple:

> Reach the exit.

But there's a problem.

The robot has:

- No map
- No instructions
- No teacher

It starts exploring.

- Takes one step forward → Hits a wall ❌
- Takes another step → Nothing happens 😐
- Moves again → Finds the correct path ⭐

The robot receives a **reward**.

Over time, it learns which actions lead to success.

This is **Reinforcement Learning**.

---

# 🔹 What Is Reinforcement Learning?

**Reinforcement Learning (RL)** is a type of Machine Learning where an **agent learns by interacting with its environment**.

Instead of learning from labeled data, the agent learns through:

- Trial and error
- Rewards
- Punishments

The goal is to discover the best strategy for maximizing rewards.

---

# 🧠 Basic Idea

```text
Agent
   ↓
Take Action
   ↓
Environment
   ↓
Reward / Penalty
   ↓
Learn Better Strategy
```

The cycle repeats until the agent learns the best behavior.

---

# 🔹 Why Is It Called "Reinforcement" Learning?

Every action taken by the agent receives feedback.

The feedback can be:

- ✅ Positive Reward
- ❌ Negative Reward (Penalty)

The agent learns by reinforcing good actions and avoiding bad ones.

# 🤖 How Reinforcement Learning Works

```text
Environment
      ↑
      │
Reward
      │
Agent ───► Action
```

The agent continuously:

1. Observes the environment
2. Takes an action
3. Receives a reward
4. Learns from the result
5. Repeats the process

---

# 🔑 Key Terms

## 1. Agent

The learner or decision-maker.

Examples:

- Robot
- Self-driving car
- Game-playing AI

---

## 2. Environment

Everything around the agent.

Examples:

- Maze
- Road
- Chess board
- Video game

---

## 3. Action

Anything the agent chooses to do.

Examples:

- Move left
- Move right
- Accelerate
- Brake
- Jump

---

## 4. Reward

Feedback received after taking an action.

Examples:

| Action | Reward |
|----------|---------|
| Reach goal | +100 |
| Hit wall | -10 |
| Fall into hole | -50 |
| Collect coin | +5 |

Rewards help the agent understand which actions are good.

---

## 5. Policy

A **policy** is the strategy the agent develops over time.

It answers the question:

> "What should I do in this situation?"

The better the policy, the better the agent performs.

---

# 🎯 Goal of Reinforcement Learning

The objective is simple:

> **Maximize the total reward over time.**

The agent does not try to win one step at a time.

Instead, it learns the sequence of actions that leads to the highest long-term reward.

---

# 🌟 AlphaGo: A Famous Example

One of the greatest achievements in Reinforcement Learning was **AlphaGo**, developed by Google DeepMind.

---

## What Is Go?

Go is an ancient strategy board game.

Compared to chess:

- More possible moves
- Greater complexity
- Extremely difficult for computers

---

## How Did AlphaGo Learn?

Instead of only studying human games, AlphaGo:

- Played millions of games against itself.
- Learned from victories and defeats.
- Improved its strategy continuously.

---

## Historic Achievement

In **2016**, AlphaGo defeated the world Go champion.

This demonstrated the power of Reinforcement Learning.

# 🔄 Reinforcement Learning Workflow

```text
Observe Environment
         ↓
     Choose Action
         ↓
 Receive Reward/Penalty
         ↓
 Update Strategy (Policy)
         ↓
 Repeat
```

# ⚖️ Comparison of Machine Learning Types

| Feature | Supervised Learning | Unsupervised Learning | Reinforcement Learning |
|----------|---------------------|-----------------------|------------------------|
| Uses Labels | ✅ Yes | ❌ No | ❌ No |
| Teacher Available | ✅ Yes | ❌ No | ❌ No |
| Learns Through | Labeled Data | Hidden Patterns | Rewards & Penalties |
| Goal | Predict Labels | Discover Patterns | Maximize Reward |
| Example | Spam Detection | Customer Segmentation | Self-Driving Cars |
