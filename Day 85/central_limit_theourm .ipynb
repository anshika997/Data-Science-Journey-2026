# Central Limit Theorem (CLT)

---

## 1. What is the Mean?

The **mean** (or average) of a dataset is the sum of all values divided by the number of values.

### Formula

$$
\text{Mean} = \frac{x_1 + x_2 + x_3 + \cdots + x_n}{n}
$$

It represents the central value of the data.

---

## 2. What is the Standard Deviation?

The **standard deviation** measures how spread out the numbers are from the mean.

### Steps to Calculate Standard Deviation

1. Find the mean
2. Subtract the mean from each value
3. Square each difference
4. Calculate the average of these squared differences (variance)
5. Take the square root of the variance

### Formula

$$
\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i-\mu)^2}
$$

Where:

- \( \mu \) = Mean
- \( \sigma \) = Standard Deviation

### Interpretation

- Smaller standard deviation → Data points are close to the mean
- Larger standard deviation → Data is more spread out

---

## 3. Central Limit Theorem (CLT)

The **Central Limit Theorem (CLT)** states:

> If you take many random samples of size \(n\) from any population (with finite mean and variance), then the distribution of the sample means will tend to be approximately normal as \(n\) becomes large, regardless of the shape of the original population.

This is one of the most important results in statistics.

---

## Why is CLT Important?

Even if the original population is:

- Skewed
- Uniform
- Randomly shaped

The distribution of sample means becomes approximately **Normal (Bell-Shaped)** when the sample size is sufficiently large.

This allows us to use normal distribution techniques for inference and prediction.

---

## 4. Mathematical Expression

Let:

$$
X_1, X_2, \ldots, X_n
$$

be \(n\) independent and identically distributed (i.i.d.) random variables with:

- Mean = \( \mu \)
- Standard Deviation = \( \sigma \)

Then the sample mean:

$$
\bar{X} = \frac{X_1 + X_2 + \cdots + X_n}{n}
$$

approaches a normal distribution as \(n\) becomes large.

---

## Sampling Distribution of the Mean

The sampling distribution of \( \bar{X} \) has:

### Mean

$$
\mu_{\bar{X}} = \mu
$$

### Standard Error

$$
\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}
$$

Where:

- \( \sigma \) = Population standard deviation
- \( n \) = Sample size

---

## Key Insight

As the sample size increases:

$$
\frac{\sigma}{\sqrt{n}}
$$

becomes smaller.

This means:

- Sample means cluster more tightly around the true population mean
- Estimates become more reliable

---

## Example

Suppose:

- Population Mean = 100
- Population Standard Deviation = 20
- Sample Size = 25

Then:

### Mean of Sample Means

$$
\mu_{\bar{X}} = 100
$$

### Standard Error

$$
\sigma_{\bar{X}} = \frac{20}{\sqrt{25}}
= \frac{20}{5}
= 4
$$

So the sampling distribution of the sample mean is approximately:

$$
N(100, 4)
$$

---

## Applications of CLT in Data Science

- Statistical inference
- Confidence intervals
- Hypothesis testing
- Machine learning evaluation
- A/B testing
- Survey analysis

---

## Summary

| Concept | Formula |
|----------|----------|
| Mean | \( \mu = \frac{\sum x_i}{n} \) |
| Standard Deviation | \( \sigma = \sqrt{\frac{1}{n}\sum (x_i-\mu)^2} \) |
| Sample Mean | \( \bar{X} = \frac{\sum X_i}{n} \) |
| Mean of Sample Means | \( \mu_{\bar{X}} = \mu \) |
| Standard Error | \( \sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}} \) |
| CLT Result | Sample means become approximately normal as \(n\) increases |

---

## 📌 Key Takeaway

The Central Limit Theorem explains why the **Normal Distribution** appears so frequently in statistics and data science. No matter the original population shape, sufficiently large random samples produce sample means that follow a normal distribution.
