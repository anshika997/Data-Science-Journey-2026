# Basic Rules of Probability

---

## 1. Important Definitions

Understanding these basic definitions is crucial for learning probability.

- **Experiment**: An action or process that leads to one or more outcomes  
  - Deterministic → predictable outcome  
  - Probabilistic (Random) → uncertain outcome  

- **Outcome**: A possible result of an experiment  

- **Probability (P)**: Likelihood that an event occurs (ranges from 0 to 1)  

- **P(A ∪ B)**: Probability of event A or event B occurring  

- **P(A ∩ B)**: Probability of both A and B occurring  

- **Sample Space (S)**: Set of all possible outcomes  

- **Event (E)**: Subset of the sample space  

- **Mutually Exclusive Events**: Cannot occur together  
  ```
  P(A ∩ B) = 0
  ```

- **Independent Events**: One event does not affect the other  
  ```
  P(A ∩ B) = P(A) × P(B)
  ```

- **Certain Event**: Probability = 1  

- **Impossible Event**: Probability = 0  

- **Exhaustive Events**: Cover the entire sample space  

### Example

Rolling a die:

- Sample Space: `S = {1, 2, 3, 4, 5, 6}`  
- Event (Even numbers): `{2, 4, 6}`  

---

## 2. Complement Rule

The complement of an event A is the event that A does not occur.

- Notation: `Aᶜ`

### Formula

$$
P(A^c) = 1 - P(A)
$$

### Example

If probability of rain = 0.3:

$$
P(\text{No Rain}) = 1 - 0.3 = 0.7
$$

---

## 3. Addition Rule

Used to calculate probability of union of events.

### General Rule

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

### For Mutually Exclusive Events

$$
P(A \cup B) = P(A) + P(B)
$$

### Example

Given:
- P(A) = 0.4  
- P(B) = 0.5  
- P(A ∩ B) = 0.2  

$$
P(A \cup B) = 0.4 + 0.5 - 0.2 = 0.7
$$

---

## 4. Multiplication Rule

Used to calculate probability of both events occurring.

### For Independent Events

$$
P(A \cap B) = P(A) \times P(B)
$$

### Example

- Event A: rolling a 3 → P(A) = 1/6  
- Event B: rolling an even number → P(B) = 3/6  

$$
P(A \cap B) = \frac{1}{6} \times \frac{3}{6} = \frac{1}{12}
$$

---

### For Dependent Events

$$
P(A \cap B) = P(A) \times P(B|A)
$$

### Example

A bag has 5 red and 3 blue balls.

- P(A) = 5/8 (first red)  
- P(B|A) = 4/7 (second red after one removed)  

$$
P(A \cap B) = \frac{5}{8} \times \frac{4}{7} = \frac{20}{56} = \frac{5}{14}
$$

---

## 5. Independent vs Dependent Events

- **Independent**: One event does not affect the other  
  - Example: Tossing two coins  

- **Dependent**: One event affects the other  
  - Example: Drawing cards without replacement  

---

## 6. Mutually Exclusive Events

Two events are mutually exclusive if they cannot happen at the same time.

### Example

- Event A: Rolling a 2  
- Event B: Rolling a 5  

These cannot occur together.
