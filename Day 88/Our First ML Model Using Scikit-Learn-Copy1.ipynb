# 🔹 Step 2: Build Your First Machine Learning Model

## 🎯 Problem Statement

We have some fruits.

### Apples
- Heavy
- Red

### Oranges
- Lighter
- Orange-colored

Instead of writing many `if-else` rules, we'll train a machine to learn the pattern automatically.

---

# 🔹 Training Data

### Features

We will use:

1. Weight (grams)
2. Color
   - `0 = Red`
   - `1 = Orange`

### Labels

- Apple
- Orange

---

# 🔹 Code Example

```python
from sklearn import tree

# Step 1: Collect Data

# Features: [weight, color]
features = [
    [150, 0],  # Apple
    [170, 0],  # Apple
    [130, 1],  # Orange
    [120, 1]   # Orange
]

# Labels
labels = [
    "apple",
    "apple",
    "orange",
    "orange"
]

# Step 2: Train the Model
clf = tree.DecisionTreeClassifier()

clf = clf.fit(features, labels)

# Step 3: Make Predictions

print(clf.predict([[160, 0]]))
print(clf.predict([[115, 1]]))
```

# 🌳 Visualize the Decision Tree

We can see how the model makes decisions.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plot_tree(
    clf,
    feature_names=["weight", "color"],
    class_names=["apple", "orange"],
    filled=True
)

plt.show()
```

# 🔹 What is a Decision Tree?

A Decision Tree makes predictions by asking questions such as:

```text
Is color = Red?

├── Yes → Apple
└── No  → Orange
```

The tree learns these decision rules automatically from data.
