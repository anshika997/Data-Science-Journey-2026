# Uniform Distribution

---

## What It Is

A **uniform distribution** is when **every outcome is equally likely**.

---

## Simple Example: Rolling a Fair Die

- Possible outcomes: `{1, 2, 3, 4, 5, 6}`  
- Each outcome has probability: `1/6`  

This is a **discrete uniform distribution**.

---

## Continuous Version

If we randomly pick a number between **0 and 1**:

- Every value in this range is equally likely  

This is a **continuous uniform distribution**.

---

## Graphs to Visualize

### 1. Discrete Uniform (Die Roll)

```
Outcome:     1   2   3   4   5   6
Probability: |---|---|---|---|---|---|
              1/6 for each (flat bars)
```

---

### Discrete Uniform Formula

$$
P(X = x) = \frac{1}{n}
$$

Where:
- \( n \) = number of possible outcomes  

**Example:**
- For a 6-sided die → \( P(4) = 1/6 \)

---

### 2. Continuous Uniform (0 to 1)

- Represented by a **flat horizontal line**  
- Probability density is constant across the interval  

---

### Continuous Uniform Formula

$$
f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
$$

- Outside the range \([a, b]\), probability = 0  

---

## Why It’s Useful in Data Science

- Models pure randomness  
- Used in simulations  
- Used in generating random numbers  
- Common in sampling methods  

---

## In Code (Python)

```python
import numpy as np
import matplotlib.pyplot as plt

# Continuous uniform from 0 to 1
samples = np.random.uniform(0, 1, 10000)

plt.hist(samples, bins=50, density=True, alpha=0.6)
plt.title("Continuous Uniform Distribution (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

You’ll see a **flat histogram**, showing equal probability across values.

## 📌 Key Insight

Uniform distribution represents **complete randomness** — every outcome has equal chance.

````python
import numpy as np
import matplotlib.pyplot as plt

# Continuous uniform from 0 to 1
samples = np.random.uniform(0, 1, 10000)

plt.hist(samples, bins=50, density=True, alpha=0.6)
plt.title("Continuous Uniform Distribution (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
````
