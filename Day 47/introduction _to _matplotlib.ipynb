## Real Data Example: Sachin Tendulkar’s Runs Over Time

```python
years = [1990 , 1992 , 1994 , 1996 , 1998 , 2000 , 2003 , 2005 , 2007 , 2010]
runs = [500 , 700 , 1100 , 1500 , 1800 , 1200 , 1700 , 1300 ,900 ,1500] 
plt.plot(years,runs) # years x axis runs y axis
# plt.show()
```

## Adding X and Y Labels

```python
plt.plot(years,runs)
plt.xlabel(" Years ")
plt.ylabel(" Runs Scored ")
plt.title(" Sachin Tendulkar’s Runs Over Time ")
```

## Multiple Lines in One Plot

```python
kohli = [0,0,500,800,1100,1300,1500,1800,1900,2100]
sehwag = [0,300,800,1200,1500,1700,1600,1400,1000,0]
plt.plot(years,kohli,label= " KING VIRAT KOHLI ") # THIS IS BETTER WAY
# or we can use
# plt.legend(["KING VIRAT KOHLI ","SEHWAG"]) BAD PRACTICE 
plt.plot(years,sehwag,label= " SEHWAG ")
plt.xlabel("YEARS")
plt.ylabel("RUNS SCORED ")
plt.title("PERFORMANCE COMPARISON ")
plt.legend()
plt.show()


```

## Using Format Strings
it is not readable 

```python
plt.plot(years,kohli,'bo--',label = " KING VIRAT KOHLI ")# blue circle with dotted line
plt.plot(years,sehwag,'g^:',label = " SEHWAG ")# green triangle with dotts
plt.legend()
```

## Color and Line Style Arguments

```python
plt.plot(years,kohli,color='blue',linestyle='--',linewidth=4,label="KING VIRAT KOHLI")
plt.plot(years,sehwag,color='pink',linestyle='-.',label="SEHWAG")
plt.plot(years,runs,color='green',label="TENDULKAR")
plt.tight_layout()
plt.grid(True)
plt.style.use('fast')

# plt.style.use('classic')
# plt.style.use('_classic_test_patch')
# plt.style.use('dark_background')
# plt.style.use('ggplot')
# plt.style.use('grayscale')



plt.legend()

```

## Visualizing Tons of Data – What Crowded Looks Like

```python
import matplotlib
import numpy as np 
for i in range(50):
    plt.plot(np.random.rand(100),linewidth=3)
plt.title("too much data can confusing")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Final Tips for Beginners

- Always start with simple plots
- Add labels and legends early
- Use plt.grid() and plt.tight_layout() to improve readability
- Try different styles to find what works for your use case
