Object-Oriented Programming (OOP) is a programming paradigm that organizes code into objects that contain both data (attributes) and behavior (methods).

```python
class Employee:
    company="hp"
    def printDetails(self):
        print(f"the company is {self.company} ")
e = Employee()
e.printDetails()
```

```python
class Employee:
    company="hp"
    
    def __init__(self,name,age,salary):
        
        self.name=name
        self.age=age
        self.salary=salary

    
    def printDetails(self):
        print(f"the company is {self.company} and the age of the employee is {self.age} and name is {self.name} with the salary of ${self.salary}")
e = Employee("anshika",45,45000)
e.printDetails()
```

```python
class Employee:
    company="hp"
    
    def __init__(self,name,age,salary):
        
        self.name=name
        self.age=age
        self.salary=salary

    
    def printDetails(self):
        print(f"the company is {self.company} and the age of the employee is {self.age} and name is {self.name} with the salary of ${self.salary}")
    @staticmethod
    def printTime():# here it runs independently without self or ye self method ka kuch bhi use nhi karega na name na company na salary 
        print("the time is now ")

    @classmethod
    def printClassDetails(cls):
        print(f"the company is {cls.company}")
        #print(f"the name  is {cls.name}")# its not a class atribute its a instance variable so it only run with class 

e = Employee("anshika",45,45000)
e.printDetails()
e.printTime()
e.printClassDetails()
```
