# 🗄️ Stored Procedures in MySQL

A **Stored Procedure** is a saved block of SQL code that you can execute later by calling its name.

👉 It allows you to group SQL statements and reuse them—just like a function in programming.

---

# 🔹 Why Use Stored Procedures?

* Avoid repeating the same SQL logic
* Improve performance by reducing network traffic
* Encapsulate complex business logic inside the database

---

# 🔹 Creating a Stored Procedure

When creating a stored procedure, you must temporarily change the SQL statement delimiter from `;` to something else like `//` or `$$`.

---

## ❓ Why Change the DELIMITER?

MySQL ends a command at the first `;`.

Since stored procedures contain multiple SQL statements (each ending in `;`), we need to prevent MySQL from ending the procedure too early.

👉 So we:

1. Change delimiter
2. Write procedure
3. Change it back

---

## 🔹 Example: Simple Procedure to List All Employees

```sql id="y9r7vl"
DELIMITER //

CREATE PROCEDURE list_employees()
BEGIN
    SELECT * FROM employees;
END //

DELIMITER ;
```

✔ This creates a procedure named `list_employees`.

---

# 🔹 Calling a Stored Procedure

Use the `CALL` statement:

```sql id="gx3y7y"
CALL list_employees();
```

---

# 🔹 Stored Procedure with Parameters

You can pass values into procedures using the `IN` keyword.

---

## 🔹 Example: Get Employee by ID

```sql id="1y0l2p"
DELIMITER //

CREATE PROCEDURE get_employee_by_id(IN emp_id INT)
BEGIN
    SELECT * FROM employees WHERE id = emp_id;
END //

DELIMITER ;
```

✔ `IN emp_id INT` means:
Take an integer input called `emp_id` when this procedure is called.

---

## 🔹 Calling with Parameter

```sql id="8s6n2k"
CALL get_employee_by_id(3);
```

---

# 🔹 Dropping a Stored Procedure

To delete a stored procedure:

```sql id="l0z9qp"
DROP PROCEDURE IF EXISTS get_employee_by_id;
```

✔ Prevents error if the procedure doesn’t exist.

---

# 📊 Summary

| Task             | SQL Command                                |
| ---------------- | ------------------------------------------ |
| Create Procedure | `CREATE PROCEDURE`                         |
| Change Delimiter | `DELIMITER //`                             |
| Call Procedure   | `CALL procedure_name();`                   |
| Input Parameter  | `IN param_name data_type`                  |
| Drop Procedure   | `DROP PROCEDURE IF EXISTS procedure_name;` |

---

# ✅ Best Practices

* Use clear and meaningful names
* Use `IN`, `OUT`, or `INOUT` for flexibility
* Keep business logic in DB only if it improves clarity/performance

---
