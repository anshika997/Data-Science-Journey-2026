# Conditional Probability

---

## What is Conditional Probability?

Conditional probability is the probability of an event **A** occurring, given that another event **B** has already occurred. It helps us update our understanding based on new information.

### Formula

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

This is read as:  
> "The probability of A given B."

---

## Law of Total Probability

The Law of Total Probability helps you find the probability of an event **A** by breaking it down using other known events \( B_1, B_2, ..., B_n \).

### Formula

$$
P(A) = P(A|B_1) \cdot P(B_1) + P(A|B_2) \cdot P(B_2) + \dots + P(A|B_n) \cdot P(B_n)
$$

---

## Conditions to Use This Law

### 1. Partition of the Sample Space

The events \( B_1, B_2, ..., B_n \) must:

- **Mutually exclusive** (no overlap)  
- **Collectively exhaustive** (cover the entire sample space)  

---

### 2. Known Probabilities

You must know:

- \( P(B_i) \) for each event  
- \( P(A|B_i) \)  

---

### 3. Non-zero Probability

$$
P(B_i) > 0
$$

You cannot condition on an impossible event.

---

## Real-life Analogy

A box contains:
- 5 red balls  
- 3 blue balls  
- 2 green balls  

You are told the ball picked is **not green**. What is the probability it is red?

Let:
- \( A \) = ball is red  
- \( B \) = ball is not green  

Out of 8 non-green balls (5 red + 3 blue), 5 are red.

$$
P(\text{Red}|\text{Not Green}) = \frac{5}{8}
$$

---

## Formula Breakdown

If we know:
- \( P(A \cap B) \): probability both A and B occur  
- \( P(B) \): probability B occurs  

Then:

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

This means:  
> Among all outcomes where B occurs, how many also include A?

---

## Example Problem

A class has:
- 60% boys → \( P(B) = 0.6 \)  
- 40% girls → \( P(G) = 0.4 \)  

- 30% of boys play football → \( P(F|B) = 0.3 \)  
- 10% of girls play football → \( P(F|G) = 0.1 \)  

### Step 1: Total probability of playing football

$$
P(F) = P(F|B)\cdot P(B) + P(F|G)\cdot P(G)
$$

$$
P(F) = 0.3 \cdot 0.6 + 0.1 \cdot 0.4 = 0.18 + 0.04 = 0.22
$$

---

### Step 2: Find probability student is a boy given they play football

$$
P(B|F) = \frac{P(F|B)\cdot P(B)}{P(F)}
$$

$$
P(B|F) = \frac{0.18}{0.22} \approx 0.818
$$

So, there is about an **81.8% chance** that a football player is a boy.

---

## ⚠️ Common Mistakes to Avoid

- Confusing \( P(A|B) \) with \( P(B|A) \)  
- Forgetting to divide by \( P(B) \)  
- Assuming both probabilities are the same (they are NOT)  

---

## ✅ Summary

- Conditional Probability:  
  $$
  P(A|B) = \frac{P(A \cap B)}{P(B)}
  $$

- Law of Total Probability helps break complex problems  
- Always check conditions (mutual exclusivity, completeness)  
- Be careful with interpretation of conditional probabilities
