# 🔹 What Is Unsupervised Learning?

Unsupervised Learning is a type of Machine Learning where the model learns from **unlabeled data**.

Unlike supervised learning:

- There are no correct answers.
- No teacher is available.
- No labels are provided.

The machine receives only the data and tries to discover hidden patterns on its own.

---

# 🎯 Basic Idea

```text
Unlabeled Data
       ↓
   Learning
       ↓
 Pattern Discovery
       ↓
 Groups / Insights
```

The goal is not prediction.

The goal is **exploration and pattern discovery**.

---

# 🔹 Why Is It Called "Unsupervised"?

Because there is no supervision.

In supervised learning:

```text
Input → Correct Answer
```

In unsupervised learning:

```text
Input → ?
```

The machine must figure out relationships without knowing the answers.

# 🛒 Real-World Example: Customer Segmentation

Suppose you own an online store.

You have information about customers:

- Products purchased
- Visit frequency
- Spending habits
- Browsing behavior

But you do not know their customer type.

---

## Customer Data

| Customer | Visits | Spending |
|------------|---------|-----------|
| A | High | High |
| B | Low | Low |
| C | High | Medium |

An unsupervised learning algorithm can automatically create groups.

---

## Possible Groups

### 💰 Big Spenders

- Spend a lot of money
- Purchase frequently

### 🏷️ Bargain Hunters

- Buy mostly discounted items
- Spend less

### 🌙 Night-Time Browsers

- Visit the website mainly at night

These labels were not provided.

The machine discovered the groups itself.

---

Businesses can:

- Send targeted advertisements

- Increase sales

Example:

```text
Big Spenders
→ Premium Loyalty Program

Bargain Hunters
→ Discount Coupons
```

# 🔹 Common Unsupervised Learning Algorithms

You don't need to memorize all of them, but it's useful to know the major ones.

---

## 1. K-Means Clustering

Most popular clustering algorithm.

### Goal

Group similar data points into clusters.

Example:

```text
Customers
      ↓
 K-Means
      ↓

Cluster 1 → Big Spenders
Cluster 2 → Bargain Hunters
Cluster 3 → Occasional Buyers
```

---

## 2. Hierarchical Clustering

Creates a hierarchy of groups.

Think of it as building a family tree of data.

Example:

```text
Animals
├── Mammals
│   ├── Cats
│   └── Dogs
└── Birds
```

---

## 3. PCA (Principal Component Analysis)

Used for data compression and visualization.

### Purpose

Reduce large datasets into fewer important features.

Example:

```text
100 Features
      ↓
     PCA
      ↓
  5 Important Features
```

Benefits:

- Faster processing
- Easier visualization
- Reduced storage requirements

# ⚠️ Challenges of Unsupervised Learning

Since there are no correct answers:

- Results may be harder to evaluate.
- Clusters may not always make sense.
- Different algorithms may produce different patterns.

However, it remains extremely useful for data exploration.

---

# 🔄 Supervised vs Unsupervised Learning

| Feature | Supervised Learning | Unsupervised Learning |
|----------|--------------------|----------------------|
| Labels Available | ✅ Yes | ❌ No |
| Teacher Present | ✅ Yes | ❌ No |
| Goal | Predict labels | Discover patterns |
| Output | Predictions | Groups/Insights |
| Example | Spam Detection | Customer Segmentation |

---

# 📋 Summary

- Common tasks include clustering and dimensionality reduction.
- Customer segmentation is a popular application.
- K-Means, Hierarchical Clustering, and PCA are common algorithms.
